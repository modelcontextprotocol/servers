## üìö MCP Server Quick Start Guide

This tutorial provides a quick overview of your generated Model Context Protocol (MCP) server, how it works, and how to use the generated files.

### What is an MCP Server?

An **MCP Server** acts as an intermediary layer that exposes an external API's functionalities (tools, resources, prompts) in a standardized format that Large Language Models (LLMs) can understand and interact with. It allows LLMs to use external services (like GitHub, Slack, your custom API) by calling defined 'tools'.

Your generated server is configured for the `example_service` using `TypeScript` and `stdio` transport, authenticating via an `API Key`, and exposes one tool: `example_service_get_status`.

### What Your Generated Server Does

Your specific MCP server is designed to:
1.  **Expose the `example_service_get_status` tool:** This tool is defined to retrieve the status of a service at the `/status` endpoint using a `GET` request. It accepts `service_id` (required) and `detail_level` (optional) parameters.
2.  **Authenticate using an API Key:** It expects an API key to be provided via an environment variable (`MY_SERVICE_API_KEY`) and passed in the `X-API-Key` HTTP header when making calls to the `https://api.example.com/v1` API.
3.  **Use `stdio` transport:** This means it communicates over standard input/output, which is common for local or desktop applications interacting with an LLM.

2.  **Install Dependencies:**

3.  **Environment Variables:** Set the `MY_SERVICE_API_KEY` environment variable with your actual API key before running the server.
    *   Example (Linux/macOS): `export MY_SERVICE_API_KEY="your_actual_api_key_here"`
    *   Example (Windows PowerShell): `$env:MY_SERVICE_API_KEY="your_actual_api_key_here"`
4.  **Run the Server:**
    *   For TypeScript: `npm start` (or `node dist/server.js` if it compiles to `dist/server.js`)
    *   For Python: `python src/server.py`
5.  **Integrate with LLM:** Your running MCP server can now be connected to an LLM via the MCP SDK (TypeScript or Python) to allow the LLM to call your defined tools.

# üîµ Server Configuration

Basic information about your MCP server.

```python
# Server configuration data structure
server_config = {
    'service_name': '',
    'description': '',
    'version': '1.0.0',
    'author': '',
    'language': 'typescript',  # typescript or python
    'transport': 'stdio',  # stdio or http
}

# Service name
service_name = widgets.Text(
    description='Service Name:',
    placeholder='e.g., github, slack, jira',
    style={'description_width': '150px'},
    layout=widgets.Layout(width='500px')
)
service_name_help = widgets.HTML(
    value='<small><i>Name of the service you\'re integrating (lowercase, no spaces)</i></small>'
)

# Description
description = widgets.Textarea(
    description='Description:',
    placeholder='Brief description of what your MCP server does',
    style={'description_width': '150px'},
    layout=widgets.Layout(width='500px', height='80px')
)

# Version
version = widgets.Text(
    description='Version:',
    value='1.0.0',
    style={'description_width': '150px'},
    layout=widgets.Layout(width='200px')
)

# Author
author = widgets.Text(
    description='Author:',
    placeholder='Your name or organization',
    style={'description_width': '150px'},
    layout=widgets.Layout(width='400px')
)

# Language selection
language = widgets.RadioButtons(
    options=['TypeScript (recommended)', 'Python'],
    description='Language:',
    style={'description_width': '150px'}
)
language_help = widgets.HTML(
    value='<small><i>TypeScript is recommended for better type safety and wider compatibility</i></small>'
)

# Transport selection
transport = widgets.RadioButtons(
    options=['stdio (local/desktop apps)', 'HTTP (web services)'],
    description='Transport:',
    style={'description_width': '150px'}
)
transport_help = widgets.HTML(
    value='<small><i>stdio for local integrations, HTTP for remote/web services</i></small>'
)

def update_server_config(change):
    server_config['service_name'] = service_name.value.strip().lower().replace(' ', '-')
    server_config['description'] = description.value.strip()
    server_config['version'] = version.value.strip()
    server_config['author'] = author.value.strip()
    server_config['language'] = 'typescript' if 'TypeScript' in language.value else 'python'
    server_config['transport'] = 'stdio' if 'stdio' in transport.value else 'http'

service_name.observe(update_server_config, 'value')
description.observe(update_server_config, 'value')
version.observe(update_server_config, 'value')
author.observe(update_server_config, 'value')
language.observe(update_server_config, 'value')
transport.observe(update_server_config, 'value')

print("üîµ Server Configuration")
display(service_name, service_name_help)
display(description)
display(version)
display(author)
display(widgets.HTML("<br>"))
display(language, language_help)
display(widgets.HTML("<br>"))
display(transport, transport_help)
```

# üü¢ Authentication Configuration

How will your MCP server authenticate with the external service?

```python
# Authentication data structure
auth_config = {
    'type': 'none',  # none, api_key, oauth2, bearer
    'env_var_name': '',
    'header_name': '',
    'instructions': ''
}

# Auth type
auth_type = widgets.Dropdown(
    options=['None', 'API Key', 'Bearer Token', 'OAuth 2.0'],
    description='Auth Type:',
    style={'description_width': '150px'},
    layout=widgets.Layout(width='300px')
)

# Environment variable name
env_var = widgets.Text(
    description='Env Variable:',
    placeholder='e.g., GITHUB_TOKEN, SLACK_API_KEY',
    style={'description_width': '150px'},
    layout=widgets.Layout(width='400px')
)
env_var_help = widgets.HTML(
    value='<small><i>Environment variable name where auth credentials will be stored</i></small>'
)

# HTTP header name
header_name = widgets.Text(
    description='Header Name:',
    placeholder='e.g., Authorization, X-API-Key',
    value='Authorization',
    style={'description_width': '150px'},
    layout=widgets.Layout(width='400px')
)

# Auth instructions
auth_instructions = widgets.Textarea(
    description='Instructions:',
    placeholder='How users should obtain credentials',
    style={'description_width': '150px'},
    layout=widgets.Layout(width='500px', height='60px')
)

def update_auth_config(change):
    auth_config['type'] = auth_type.value.lower().replace(' ', '_')
    auth_config['env_var_name'] = env_var.value.strip()
    auth_config['header_name'] = header_name.value.strip()
    auth_config['instructions'] = auth_instructions.value.strip()

auth_type.observe(update_auth_config, 'value')
env_var.observe(update_auth_config, 'value')
header_name.observe(update_auth_config, 'value')
auth_instructions.observe(update_auth_config, 'value')

print("üü¢ Authentication Configuration")
display(auth_type)
display(env_var, env_var_help)
display(header_name)
display(auth_instructions)
```

Base URL and other API settings for the external service.

```python
# API configuration
api_config = {
    'base_url': '',
    'timeout': 30,
    'rate_limit': 0
}

# Base URL
base_url = widgets.Text(
    description='Base URL:',
    placeholder='https://api.example.com',
    style={'description_width': '150px'},
    layout=widgets.Layout(width='500px')
)
base_url_help = widgets.HTML(
    value='<small><i>Base URL for API requests (without trailing slash)</i></small>'
)

# Timeout
timeout = widgets.IntText(
    description='Timeout (sec):',
    value=30,
    style={'description_width': '150px'},
    layout=widgets.Layout(width='200px')
)

# Rate limit
rate_limit = widgets.IntText(
    description='Rate Limit:',
    value=0,
    style={'description_width': '150px'},
    layout=widgets.Layout(width='200px')
)
rate_limit_help = widgets.HTML(
    value='<small><i>Requests per second (0 for no limit)</i></small>'
)

def update_api_config(change):
    api_config['base_url'] = base_url.value.strip().rstrip('/')
    api_config['timeout'] = timeout.value
    api_config['rate_limit'] = rate_limit.value

base_url.observe(update_api_config, 'value')
timeout.observe(update_api_config, 'value')
rate_limit.observe(update_api_config, 'value')

print("üü° API Configuration")
display(base_url, base_url_help)
display(timeout)
display(rate_limit, rate_limit_help)
```

# üü£ Tools

Define the tools (functions) that LLMs can call through your MCP server.

**Tool Naming**: Use `{service}_{action}_{resource}` format (e.g., `github_create_issue`, `slack_send_message`)

```python
# Tools data structure
tools_list = []

# Tool input widgets
tool_name = widgets.Text(
    description='Tool Name:',
    placeholder='e.g., github_create_issue',
    style={'description_width': '120px'},
    layout=widgets.Layout(width='400px')
)

tool_description = widgets.Textarea(
    description='Description:',
    placeholder='What does this tool do?',
    style={'description_width': '120px'},
    layout=widgets.Layout(width='500px', height='80px')
)

tool_endpoint = widgets.Text(
    description='API Endpoint:',
    placeholder='/api/v1/resource',
    style={'description_width': '120px'},
    layout=widgets.Layout(width='400px')
)

tool_method = widgets.Dropdown(
    options=['GET', 'POST', 'PUT', 'PATCH', 'DELETE'],
    description='HTTP Method:',
    style={'description_width': '120px'},
    layout=widgets.Layout(width='200px')
)

tool_read_only = widgets.Checkbox(
    value=True,
    description='Read-only (safe)',
    style={'description_width': '150px'}
)

tool_destructive = widgets.Checkbox(
    value=False,
    description='Destructive (modifies data)',
    style={'description_width': '150px'}
)

# Parameters list for current tool
current_tool_params = []

# Parameter input widgets
param_name = widgets.Text(
    description='Param Name:',
    placeholder='e.g., query, user_id',
    style={'description_width': '100px'},
    layout=widgets.Layout(width='300px')
)

param_type = widgets.Dropdown(
    options=['string', 'number', 'integer', 'boolean', 'array', 'object'],
    description='Type:',
    style={'description_width': '100px'},
    layout=widgets.Layout(width='200px')
)

param_required = widgets.Checkbox(
    value=True,
    description='Required',
    style={'description_width': '100px'}
)

param_description = widgets.Text(
    description='Description:',
    placeholder='Parameter description',
    style={'description_width': '100px'},
    layout=widgets.Layout(width='400px')
)

param_default = widgets.Text(
    description='Default:',
    placeholder='Optional default value',
    style={'description_width': '100px'},
    layout=widgets.Layout(width='300px')
)

# Output areas
params_output = widgets.Output()
tools_output = widgets.Output()

def add_parameter(b):
    if not param_name.value:
        with params_output:
            clear_output()
            print("‚ö†Ô∏è Parameter name is required")
        return

    param = {
        'name': param_name.value.strip(),
        'type': param_type.value,
        'required': param_required.value,
        'description': param_description.value.strip(),
        'default': param_default.value.strip() if param_default.value else None
    }

    current_tool_params.append(param)

    # Clear inputs
    param_name.value = ''
    param_description.value = ''
    param_default.value = ''
    param_required.value = True

    # Update display
    with params_output:
        clear_output()
        print(f"‚úÖ Parameters ({len(current_tool_params)}):")
        for i, p in enumerate(current_tool_params, 1):
            req = "required" if p['required'] else "optional"
            default = f" (default: {p['default']})" if p['default'] else ""
            print(f"  {i}. {p['name']} ({p['type']}, {req}){default}")

def add_tool(b):
    if not tool_name.value or not tool_description.value:
        with tools_output:
            clear_output()
            print("‚ö†Ô∏è Tool name and description are required")
        return

    if not current_tool_params:
        with tools_output:
            clear_output()
            print("‚ö†Ô∏è Add at least one parameter first")
        return

    tool = {
        'name': tool_name.value.strip(),
        'description': tool_description.value.strip(),
        'endpoint': tool_endpoint.value.strip(),
        'method': tool_method.value,
        'parameters': current_tool_params.copy(),
        'read_only': tool_read_only.value,
        'destructive': tool_destructive.value
    }

    tools_list.append(tool)

    # Clear inputs
    tool_name.value = ''
    tool_description.value = ''
    tool_endpoint.value = ''
    tool_read_only.value = True
    tool_destructive.value = False
    current_tool_params.clear()

    # Update displays
    with params_output:
        clear_output()
        print("Parameters: (none - add parameters for next tool)")

    with tools_output:
        clear_output()
        print(f"‚úÖ Tools Added ({len(tools_list)}):")
        for i, t in enumerate(tools_list, 1):
            print(f"  {i}. {t['name']} ({t['method']} {t['endpoint']}) - {len(t['parameters'])} parameters")

# Buttons
add_param_btn = widgets.Button(
    description='‚ûï Add Parameter',
    button_style='info',
    layout=widgets.Layout(width='150px')
)
add_param_btn.on_click(add_parameter)

add_tool_btn = widgets.Button(
    description='‚úÖ Add Tool',
    button_style='success',
    layout=widgets.Layout(width='150px')
)
add_tool_btn.on_click(add_tool)

print("üü£ Tools Configuration")
print("\nüìã Step 1: Define Tool")
display(tool_name)
display(tool_description)
display(tool_endpoint)
display(tool_method)
display(widgets.HBox([tool_read_only, tool_destructive]))

print("\nüìù Step 2: Add Parameters")
display(param_name)
display(param_type)
display(param_required)
display(param_description)
display(param_default)
display(add_param_btn)
display(params_output)

print("\n‚úÖ Step 3: Save Tool")
display(add_tool_btn)
display(tools_output)
```
